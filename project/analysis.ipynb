```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#density ->	The estimated population per square kilometer.
#population -> 	An estimate of the city's urban population.
```

```python
merged_fin = pd.read_csv("Final_data.csv")
merged_fin.head(2)
```

```python
top_by_listings = merged_fin.sort_values(by='listing_count', ascending=False).head(5)
top_by_population = merged_fin.sort_values(by='population', ascending=False).head(5)
top_by_price = merged_fin.sort_values(by='price_mean', ascending=False).head(5)
top_by_area = merged_fin.sort_values(by='density', ascending=False).head(5)
```

```python
plt.figure(figsize=(10, 5))  # Set figure size to 10x5 inches and high DPI
plt.bar(top_by_population['city'], top_by_population['population'], width=0.6)

# Add labels and title
plt.xlabel('City', fontsize=10)  # Reduced font size for the labels
plt.ylabel('Listing Count', fontsize=10)  # Reduced font size for the labels
plt.title('Bar Plot for Population by City', fontsize=12)  # Adjusted title font size
plt.xticks( rotation=45)  # Smaller and rotated x-axis labels for readability
plt.yticks()  # Smaller y-axis labels

# Uncomment the line below to save the plot as an image
# plt.savefig('bar_plot_by_city.png', format='png', dpi=300)

# Show the plot
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 6))

# Plot population density as bars
ax1.bar(top_by_population['city'], top_by_population['population'], color='skyblue', label='Population')
ax1.set_xlabel('City', fontsize=12)
ax1.set_ylabel('Total Population (Exponential)', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Add a second y-axis for private rooms
ax2 = ax1.twinx()
ax2.plot(top_by_population['city'], top_by_population['listing_count'], color='red', marker='o', label='Listings')
ax2.set_ylabel('Total listings', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add title and legend
plt.title('Top Population and Total listings relationship', fontsize=14)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 6))  # High-resolution figure

# Plot population density as bars
ax1.bar(top_by_area['city'], top_by_area['density'], color='skyblue', label='Population Density')
ax1.set_xlabel('City', fontsize=12)
ax1.set_ylabel('Population Density (per km)', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Add a second y-axis for private rooms
ax2 = ax1.twinx()
ax2.plot(top_by_area['city'], top_by_area['price_mean'], color='red', marker='o', label='Mean Price')
ax2.set_ylabel('Population Density (per km) & Mean Price', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add title and legend
plt.title('Population Density and Mean Price relationship', fontsize=14)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 6))  # High-resolution figure

# Plot population density as bars
ax1.bar(top_by_area['city'], top_by_area['density'], color='skyblue', label='Population Density')
ax1.set_xlabel('City', fontsize=12)
ax1.set_ylabel('Population Density (per km)', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Add a second y-axis for private rooms
ratio = top_by_area["private_room_sum"] / top_by_area["listing_count"]
ax2 = ax1.twinx()
ax2.plot(top_by_area['city'], ratio, color='red', marker='o', label='Ratio')
ax2.set_ylabel('Total Private rooms & total population ratio', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add title and legend
plt.title('Population Density and Private Rooms relationship', fontsize=14)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```
