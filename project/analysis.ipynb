```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#density ->	The estimated population per square kilometer.
#population -> 	An estimate of the city's urban population.
```

```python
merged_fin = pd.read_csv("Final_data.csv")
merged_fin.head(2)
```

```python
top_by_listings = merged_fin.sort_values(by='listing_count', ascending=False).head(5)
top_by_population = merged_fin.sort_values(by='population', ascending=False).head(5)
top_by_price = merged_fin.sort_values(by='price_mean', ascending=False).head(5)
top_by_area = merged_fin.sort_values(by='density', ascending=False).head(5)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(top_by_population['city'], top_by_population['population'], width=0.6)

plt.xlabel('City', fontsize=10)
plt.ylabel('Listing Count', fontsize=10)
plt.title('Bar Plot for Population by City', fontsize=12)
plt.xticks( rotation=45)
plt.yticks()

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 4))

ax1.bar(top_by_population['city'], top_by_population['population'], color='skyblue', label='Population', width=0.4)
ax1.set_xlabel('City', fontsize=10)
ax1.set_ylabel('Total Population (Exponential)', fontsize=10, color='blue')
ax1.tick_params(axis='y', labelcolor='blue', labelsize=8)

ax2 = ax1.twinx()
ax2.plot(top_by_population['city'], top_by_population['listing_count'], color='red', marker='o', label='Listings')
ax2.set_ylabel('Total Listings', fontsize=10, color='red')
ax2.tick_params(axis='y', labelcolor='red', labelsize=8)

plt.title('Top Population and Total Listings Relationship', fontsize=12)
ax1.legend(loc='upper left', bbox_to_anchor=(0.6, 1), fontsize=8)
ax2.legend(loc='upper left', bbox_to_anchor=(0.8, 1), fontsize=8)

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 4))

ax1.bar(top_by_area['city'], top_by_area['density'], color='skyblue', label='Population Density', width=0.4)
ax1.set_xlabel('City', fontsize=10)
ax1.set_ylabel('Population Density (per km²)', fontsize=10, color='blue')
ax1.tick_params(axis='y', labelcolor='blue', labelsize=8)

ax2 = ax1.twinx()
ax2.plot(top_by_area['city'], top_by_area['price_mean'], color='red', marker='o', label='Mean Price')
ax2.set_ylabel('Mean Price', fontsize=10, color='red')
ax2.tick_params(axis='y', labelcolor='red', labelsize=8)

plt.title('Population Density and Mean Price Relationship', fontsize=12)
ax1.legend(loc='upper left', bbox_to_anchor=(0.5, 1), fontsize=8)
ax2.legend(loc='upper left', bbox_to_anchor=(0.8, 1), fontsize=8)

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 4))

ax1.bar(top_by_area['city'], top_by_area['density'], color='skyblue', label='Population Density', width=0.4)
ax1.set_xlabel('City', fontsize=10)
ax1.set_ylabel('Population Density (per km²)', fontsize=10, color='blue')
ax1.tick_params(axis='y', labelcolor='blue', labelsize=8)

ratio = top_by_area["private_room_sum"] / top_by_area["listing_count"]
ax2 = ax1.twinx()
ax2.plot(top_by_area['city'], ratio, color='red', marker='o', label='Ratio')
ax2.set_ylabel('Private Rooms to Listings Ratio', fontsize=10, color='red')
ax2.tick_params(axis='y', labelcolor='red', labelsize=8)

plt.title('Population Density and Private Rooms Relationship', fontsize=12)
ax1.legend(loc='upper left', bbox_to_anchor=(0.5, 1), fontsize=8)
ax2.legend(loc='upper left', bbox_to_anchor=(0.8, 1), fontsize=8)

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 4)) 

ax1.bar(top_by_area['city'], top_by_area['density'], color='skyblue', label='Population Density', width=0.4)
ax1.set_xlabel('City', fontsize=10)
ax1.set_ylabel('Population Density (per km²)', fontsize=10, color='blue')
ax1.tick_params(axis='y', labelcolor='blue', labelsize=8) 


ratio = top_by_area["entire_home_apt_sum"] / top_by_area["listing_count"]
ax2 = ax1.twinx()
ax2.plot(top_by_area['city'], ratio, color='red', marker='o', label='Ratio')
ax2.set_ylabel('Home Apartments to Listings Ratio', fontsize=10, color='red')
ax2.tick_params(axis='y', labelcolor='red', labelsize=8)

plt.title('Population Density and Home Apartments Relationship', fontsize=12)
ax1.legend(loc='upper left', bbox_to_anchor=(0.2, 1), fontsize=8)
ax2.legend(loc='upper left', bbox_to_anchor=(0.5, 1), fontsize=8)

plt.tight_layout()
plt.show()
```
